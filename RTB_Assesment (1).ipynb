# **RFM Analysis**

# Recency

```python
#recency dataframe
recency_df = df_new.groupby(by='Date', as_index=False)['Id'].max()

```

```python
recency_df.head(5)
```

```python
now = dt.date(2022,2,27)
print(now)
```

```python
recency_df['Recency'] = recency_df['Date'].apply(lambda x: (now - x).days)
recency_df.drop('Date',axis = 1,inplace=True)
recency_df.head(5)
```

# Frequency

```python
temp = df_new.copy()
temp.drop_duplicates(['invoice_num','Id'],keep='first',inplace=True)
frequency_df = temp.groupby(by=['Id'], as_index=False)['invoice_num'].count()
frequency_df.columns = ['Id','Frequency']
frequency_df.head()
```

# Monetary

```python
monetary_df = df_new.groupby(by = 'Id',as_index=False).agg({'L_quantity':'sum'})
monetary_df.columns = ['Id','L_quantity']
monetary_df.head(5)
```

```python
monetary_df.tail(5)
```

# **Create RFM Table**

```python
rfm_df = recency_df.merge(frequency_df,on='Id').merge(monetary_df,on='Id')
rfm_df.set_index('Id',inplace=True)
rfm_df.head(5)
```

# **PCA**
### **Applying PCA to reduce the the dimensions and the correlation between Frequency and Monetary features.**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import silhouette_samples,silhouette_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_roc_curve
```

```python
sc = StandardScaler()
rfm_scaled = sc.fit_transform(rfm_df)
rfm_scaled[:5]
```

```python
X = rfm_df.copy()
pca = PCA(n_components = 2)
df_pca = pca.fit_transform(X)

df_pca = pd.DataFrame(df_pca)
df_pca.head(5)
```

# **K-Means Clustering**

```python
X = df_pca.copy()
```

```python
from sklearn.cluster import KMeans 

cluster_range = range(1, 15)
cluster_errors = []
cluster_sil_scores = []

for num in cluster_range: 
    clusters = KMeans(num, n_init = 100,init='k-means++',random_state=0)
    clusters.fit(X)
    labels = clusters.labels_                     # capture the cluster lables
    centroids = clusters.cluster_centers_         # capture the centroids
    cluster_errors.append( clusters.inertia_ )    # capture the intertia
clusters_df = pd.DataFrame({ "num_clusters":cluster_range, "cluster_errors": cluster_errors} )
clusters_df[0:10]
```

```python
plt.figure(figsize=(15,6))
plt.plot(clusters_df["num_clusters"],clusters_df["cluster_errors"],marker = 'o')
plt.xlabel('count of clusters')
plt.ylabel('error')
```

```python
for num in range(2,16):
    clusters = KMeans(n_clusters=num,random_state=0)
    labels = clusters.fit_predict(df_pca)
    
    sil_avg = silhouette_score(df_pca, labels)
    print('For',num,'The Silhouette Score is =',sil_avg)
```

I observe from the elbow plot a sharp bend after the number of clusters increase by 2. Silhoutte Score is also the highest for 2 clusters.

But, there is also a significant reduce in cluster error as number of clusters increase from 2 to 4 and after 4, the reduction is not much.

So, we will choose n_clusters = 4 to properly segment our customers.

```python
kmeans = KMeans(n_clusters = 4)
kmeans = kmeans.fit(df_pca)
labels = kmeans.predict(df_pca)
centroids = kmeans.cluster_centers_

print(labels)
print()
print('Cluster Centers')
print(centroids)
```

```python
df_pca['Clusters'] = labels
df_pca.head()
```

```python
df_pca['Clusters'].value_counts()
```

```python
df_pca.boxplot(by = 'Clusters',figsize=(15,6))
plt.show()
```

now i will use the model to find no. of clusters.

```python
df_pca.head(5)
```

```python
X = df_pca[[0,1]]
Y = df_pca['Clusters']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)
lr = LogisticRegression(max_iter=1000,random_state=0)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print('Test accuracy = ', accuracy_score(y_test, y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors = 4) 
  
knn.fit(X_train, y_train) 
pred = knn.predict(X_test) 
  
# Predictions and Evaluations 
# Let's evaluate our KNN model !  
from sklearn.metrics import classification_report, confusion_matrix 
print(confusion_matrix(y_test, pred)) 
  
print(classification_report(y_test, pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)
regressor = DecisionTreeRegressor(random_state = 213)
regressor.fit(X_train,y_train)

pred = regressor.predict(X_test) 
print(confusion_matrix(y_test, pred)) 
  
print(classification_report(y_test, pred)) 
```
